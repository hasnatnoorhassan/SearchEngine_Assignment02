```python
import re
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

sample_df=pd.read_csv("sample.csv",header=None)[:40]
sample_df.columns = ['ARTICLE_ID', 'TITLE', 'FULL_TEXT']

def preprocess_text(text):
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    tokens = word_tokenize(text.lower())
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token.isalnum() and token not in stop_words]

    lemmatized_and_stemmed_tokens = [stemmer.stem(lemmatizer.lemmatize(token)) for token in tokens]
    preprocessed_text = ' '.join(lemmatized_and_stemmed_tokens)
    
    return preprocessed_text


sample_df['FULL_TEXT'] = sample_df['FULL_TEXT'].apply(preprocess_text)
```

```python
from nltk.tokenize import word_tokenize
vocabulary = set()
for text in sample_df['FULL_TEXT']:
    tokens = word_tokenize(text)
    vocabulary.update(tokens)
sorted_vocabulary = sorted(vocabulary)

print("Vocabulary size:", len(sorted_vocabulary))
print("Sample vocabulary terms:", sorted_vocabulary[1500:2000])
```

```python
vocabulary_with_ids = [(idx, term) for idx, term in enumerate(sorted_vocabulary)]
vocabulary_with_ids[:15]
```

```python
from collections import Counter
def calculate_tf(document, vocabulary):
    tokens = word_tokenize(document.lower())
    term_frequency = Counter(tokens)
    tf_representation = [(term_id, term_frequency.get(term, 0)) for term_id, term in vocabulary_with_ids]
    
    return tf_representation

document_tf = [(article_id, calculate_tf(full_text, vocabulary)) for article_id, full_text in zip(sample_df['ARTICLE_ID'], sample_df['FULL_TEXT'])]
print(document_tf[0])
```

```python
document_tf_filtered = [(article_id, [(term_id, count) for term_id, count in tf_repr if count > 0]) for article_id, tf_repr in document_tf]
print(document_tf_filtered[0])
```

```python
def calculate_idf(vocabulary, document_tf):
    document_frequency = {term_id: 0 for term_id, _ in vocabulary}

    for _, terms_list in document_tf:
        terms_set = set(term_id for term_id, _ in terms_list)
        for term_id in terms_set:
            document_frequency[term_id] += 1

    idf_representation = [(term_id, document_frequency[term_id]) for term_id in document_frequency]
    return idf_representation

idf_representation = calculate_idf(vocabulary_with_ids, document_tf_filtered)



```

```python
print(idf_representation)
```

```python

def calculate_tfidf(document_tf, idf_representation):
    tfidf_representation = []
    for article_id, tf_repr in document_tf:
        tfidf_repr = [(term_id, tf / idf) for term_id, tf in tf_repr for term_id_, idf in idf_representation if term_id == term_id_]
        tfidf_representation.append((article_id, tfidf_repr))
    
    return tfidf_representation
tfidf_representation = calculate_tfidf(document_tf_filtered, idf_representation)

print(tfidf_representation[0])
```

```python
def sparse_tfidf_vectors(tfidf_representation):
    max_term_id = max(term_id for doc_id, tfidf_repr in tfidf_representation for term_id, value in tfidf_repr)
    tfidf_vectors = {}  
    for article_id, tfidf_repr in tfidf_representation:
        sparse_vector = [] 
        for term_id, tfidf_value in tfidf_repr:
            if tfidf_value != 0:  
                sparse_vector.append((term_id, tfidf_value))
        tfidf_vectors[article_id] = sparse_vector
    return tfidf_vectors

tfidf_sparse_vectors = sparse_tfidf_vectors(tfidf_representation)
print(tfidf_sparse_vectors[1])  
```

```python

query_data = {
    'QUERY_ID': ['0'],
    'TITLE': ['query'],
    'FULL_TEXT': ['tell me something about anarchy']
}

query_df = pd.DataFrame(query_data)
query_df['FULL_TEXT'] = query_df['FULL_TEXT'].apply(preprocess_text)
query_tf = [(query_id, calculate_tf(full_text, vocabulary)) for query_id, full_text in zip(query_df['QUERY_ID'], query_df['FULL_TEXT'])]
query_tf_filtered = [(query_id, [(term_id, count) for term_id, count in tf_repr if count > 0]) for query_id, tf_repr in query_tf]
query_tfidf_representation = calculate_tfidf(query_tf_filtered, idf_representation)
query_tfidf_sparse_vectors = sparse_tfidf_vectors(query_tfidf_representation)
print(query_tfidf_sparse_vectors['0'])  
#print(query_df['FULL_TEXT'])
#print(query_tf_filtered)
# print(query_tfidf_representation)



```

```python
def filter_documents_by_query(query_vector, documents):
    filtered_documents = {}
    query_terms = set(index for index, _ in query_vector)

    for doc_id, doc_vector in documents.items():
        filtered_doc_vector = [(index, value) for index, value in doc_vector if index in query_terms]
        if filtered_doc_vector: 
            filtered_documents[doc_id] = filtered_doc_vector

    return filtered_documents


filtered_documents = filter_documents_by_query(query_tfidf_sparse_vectors['0'], tfidf_sparse_vectors)
print(filtered_documents)
```

```python
def relevance(query_vector, document_vectors):
    relevance_scores = {}
    for doc_id, doc_vector in document_vectors.items():
        relevance_score = 0.0
        for index, value in doc_vector:
            for query_index, query_value in query_vector:
                if index == query_index:
                    relevance_score += query_value * value
        relevance_scores[doc_id] = relevance_score
    return relevance_scores

relevance_scores = relevance(query_tfidf_sparse_vectors['0'], filtered_documents)
print("Relevance scores:", relevance_scores)
```

```python
sorted_relevance_scores = sorted(relevance_scores.items(), key=lambda x: x[1], reverse=True)
top_relevance_scores = sorted_relevance_scores[:5]
```
